```python
avg_hr_run = df[df['Type'] == 'Running']['Average Heart Rate (bpm)'].mean()
avg_hr_cycle = df[df['Type'] == 'Cycling']['Average Heart Rate (bpm)'].mean()
avg_hr_walk = df[df['Type'] == 'Walking']['Average Heart Rate (bpm)'].mean()
avg_hr_unicycle = df[df['Type'] == 'Unicycling']['Average Heart Rate (bpm)'].mean()
```

```python
df_run = df[df['Type'] == 'Running'].copy()
df_walk = df[df['Type'] == 'Walking'].copy()
df_cycle = df[df['Type'] == 'Cycling'].copy()
df_unicycle = df[df['Type'] == 'Unicycling'].copy()
```

```python
df_run_hr_all = [102, 130, 145, 160, 125, 175, 138, 143, 110, 120, 155, 165, 132, 140, 105]

# Heart rate zones and corresponding zone names and colors
hr_zones = [100, 125, 133, 142, 151, 173]
zone_names = ['Easy', 'Moderate', 'Hard', 'Very hard', 'Maximal']
zone_colors = ['green', 'yellow', 'orange', 'tomato', 'red']

# Create plot
fig, ax = plt.subplots()

# Plot histogram
n, bins, patches = ax.hist(df_run_hr_all, bins=hr_zones, alpha=0.5, edgecolor='black')

# Customize patches with zone colors
for i in range(len(patches)):
    patches[i].set_facecolor(zone_colors[i])

# Customize plot
ax.set(title='Distribution of Heart Rate by Training Intensity', 
       ylabel='Number of Activities',
       xlabel='Heart Rate (bpm)')
ax.xaxis.set(ticks=hr_zones)

plt.show()
```

```python
df_run_walk_cycle = pd.concat([df_run, df_walk, df_cycle])

# Columns for distance, climb, and speed
dist_climb_cols = ['Distance (km)', 'Climb (m)']
speed_col = ['Average Speed (km/h)']

# Calculating total distance and climb for each type of activity
df_totals = df_run_walk_cycle.groupby('Type')[dist_climb_cols].sum()

print('Totals for different training types:')
display(df_totals)

# Calculating summary statistics for each type of activity
df_summary = df_run_walk_cycle.groupby('Type')[dist_climb_cols + speed_col].describe()

# Combine totals with summary statistics
for col in dist_climb_cols:
    df_summary[(col, 'total')] = df_totals[col]

print('Summary statistics for different training types:')
display(df_summary)
```
